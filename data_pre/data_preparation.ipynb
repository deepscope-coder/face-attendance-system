```python
import gdown
import os

# Destination folder
save_dir = r"D:\arif\drive link for face"
os.makedirs(save_dir, exist_ok=True)

# List of Google Drive file links
links = [
    "YOUR DRIVE VIDEO LINKS"
]

# Helper function to extract file ID from link
def extract_id(drive_url):
    return drive_url.split("/d/")[1].split("/")[0]

# Download each video
for i, url in enumerate(links, start=1):
    file_id = extract_id(url)
    output = os.path.join(save_dir, f"video_{i:02d}.mp4")
    gdown.download(id=file_id, output=output, quiet=False)
```

```python
import cv2
import os
import glob
import random

# Set fixed video ID
video_id = "06"
frame_num = 1  # Global counter for all saved frames

# Input video folder
video_folder = r"D:\arif\drive link for face\shadan"
# Output image folder
output_folder = r"D:\arif\drive link for face\FaceF\shadan"
os.makedirs(output_folder, exist_ok=True)

# Get all video file paths
video_paths = glob.glob(os.path.join(video_folder, "*.mp4"))
video_paths.sort()

for video_path in video_paths:
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames < 650:
        print(f"⚠️ Video {video_path} has less than 650 frames. Skipping.")
        cap.release()
        continue

    # Select 650 random frame indices
    random_frames = sorted(random.sample(range(total_frames), 650))

    print(f"Processing {os.path.basename(video_path)} with {total_frames} frames...")

    current_frame = 0
    frame_index = 0

    while cap.isOpened() and frame_index < 650:
        ret, frame = cap.read()
        if not ret:
            break

        if current_frame == random_frames[frame_index]:
            # Save the frame using the fixed video_id and global frame_num
            frame_filename = os.path.join(output_folder, f"{video_id}_{frame_num:04d}.jpg")
            cv2.imwrite(frame_filename, frame)
            frame_num += 1
            frame_index += 1

        current_frame += 1

    cap.release()
    print(f"✅ Finished extracting 650 frames from: {video_path}")

print("✅ All videos processed with same ID.")
```
